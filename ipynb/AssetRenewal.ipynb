```python
import pandas as pd
pd.set_option('display.max_columns', None)

df1 = pd.read_pickle('pkl/asset_renewal.pkl')

df1 = df1[['NOTIF_QMNUM_Notification', 'NOTIF_STRNO_FunctionalLocation', 'Program2', 'Program', 'FYEstimate']].sort_values(by='NOTIF_STRNO_FunctionalLocation')
df1
```

```python
len(df1['NOTIF_STRNO_FunctionalLocation'].unique())
```

```python
# Getting all FLOC from [vw_Dim_FunctionalLocation] where CompanyCode = 5000 -> df2_no_length

import pyodbc

sql_query = """
    SELECT
        FLOC_TPLNR_FunctionalLocationInternalKey,
        FLOC_STRNO_FunctionalLocation,
        FLOC_PLTXT_FunctionalLocationDescr,
        FLOC_TYPTX_FunctionalLocationCategoryDescr,
        FLOC_EQART_TechnicalObjectType,
        FLOC_EARTX_TechnicalObjectTypeDescr,
        FLOC_BUKRS_CompanyCode,
        FLOC_STORT_Location,
        FLOC_STORT_LocationDescr
    FROM [myANALYTICS_SP].[silver.dimension.enterprise.asset].[vw_Dim_FunctionalLocation]
	WHERE FLOC_BUKRS_CompanyCode = '5000'
	ORDER BY FLOC_STRNO_FunctionalLocation
"""

# Define your server name
server_name = 'myanalytics.aurizon.com.au'

# Establish a connection using Windows Authentication
conn = pyodbc.connect('DRIVER={SQL Server};SERVER=' + server_name + ';Trusted_Connection=yes;')

# Execute the SQL query and load the result into a pandas DataFrame
df2 = pd.read_sql_query(sql_query, conn)

# Display the DataFrame
df2
```

```python
df2 = df2[['FLOC_STRNO_FunctionalLocation', 'FLOC_EARTX_TechnicalObjectTypeDescr']].sort_values(by='FLOC_STRNO_FunctionalLocation')
df2
```

```python
merged_df = pd.merge(df1, df2, how='left', left_on='NOTIF_STRNO_FunctionalLocation', right_on='FLOC_STRNO_FunctionalLocation')

merged_df = merged_df[['FLOC_EARTX_TechnicalObjectTypeDescr', 'FYEstimate']].sort_values(by='FLOC_EARTX_TechnicalObjectTypeDescr')
merged_df['FLOC_EARTX_TechnicalObjectTypeDescr'] = merged_df['FLOC_EARTX_TechnicalObjectTypeDescr'].astype(str)
merged_df['FYEstimate'] = merged_df['FYEstimate'].astype(float)
merged_df
```

```python
average_renewal_cost = merged_df.groupby('FLOC_EARTX_TechnicalObjectTypeDescr')['FYEstimate'].mean().reset_index()
pd.options.display.float_format = '{:.2f}'.format

average_renewal_cost
```

```python
file_path = 'csv/average_renewal_cost.csv'
average_renewal_cost.to_csv(file_path, index=True)
```
