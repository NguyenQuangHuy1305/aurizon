```python
import os
import pandas as pd
```

```python
folder = r'C:\Users\R893859\Aurizon Operations Limited\Corridor Asset Management Plans - FY25 Tonnage Scenario_Board Papers\NSAP Model'
files = [folder + '\\' + file for file in os.listdir(folder) if 'Tonnage Model' in file]
files
```

```python
tonnages = []
for file in files:
    t = pd.read_excel(file,sheet_name='MNT',header=8)
    t['file'] = file

    if 'base' in file.lower():
        t['Scenario'] = 'base'
    elif 'carbon constrained' in file.lower():
        t['Scenario'] = 'carbon constrained'
    elif 'carbon shock' in file.lower():
        t['Scenario'] = 'carbon shock'

    if 'blackwater' in file.lower():
        t['System'] = 'blackwater'
    elif 'goonyella' in file.lower():
        t['System'] = 'goonyella'
    elif 'moura' in file.lower():
        t['System'] = 'moura'
    elif 'newlands' in file.lower():
        t['System'] = 'newlands'
    tonnages.append(t)
t = pd.concat(tonnages,axis=0)

t.drop(columns=['ZONE', 'ORDER', 'file'], inplace=True)
t.rename(columns={i:'LineCode' for i in t.columns if 'LINECODE' in str(i)},inplace=True)
t.rename(columns={'DESCRIPTION': 'Description'}, inplace=True)

t
```

```python
numeric_column_names = []
string_column_names = []

for col in t.columns:
    try:
        int(col)
        numeric_column_names.append(col)
    except ValueError:
        string_column_names.append(col)

numeric_column_names.sort()

all_col = string_column_names + numeric_column_names
t[all_col]
```

```python
t[t['Scenario'] == 'base'][numeric_column_names].sum(axis=0)
```

```python
unpivoted_df = pd.melt(t, id_vars=['System', 'Scenario', 'Description', 'LineCode'], var_name='Year', value_name='MNT')
unpivoted_df
```
