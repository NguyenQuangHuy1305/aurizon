```python
import pandas as pd

df1 = pd.read_pickle('pkl/FAR.pkl')
df1 = df1[df1['Functional Loc.'] != '#']

df1
```

```python
df1 = df1[['Asset Number', 'LTD Cost', 'Functional Loc.']]
df1
```

```python
# grouped_df = df1.groupby('Asset Number')['LTD Cost'].sum().reset_index()

grouped_df = df1.groupby('Functional Loc.').agg({
    'LTD Cost': 'sum'  # Sum 'LTD Cost' column
}).reset_index()

grouped_df.to_pickle("pkl/asset_renewal_cost_2024_03_21.pkl")

grouped_df
```
