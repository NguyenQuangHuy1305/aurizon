```python
# Function to split the 'LineCode' values into two rows for 'CL' codes
def split_row(row):
    if '\n' in row['LineCode']:
        return row['LineCode'].split('\n')
    return row['LineCode']

# # Function to extract the last 3 characters for 'NP' codes
# def extract_last_3_numbers(row):
#     return row['LineCode'][-3:]
```

```python
df2['LineCode'] = df2.apply(lambda row: split_row(row), axis=1)
df2 = df2.explode('LineCode')

df2
```
