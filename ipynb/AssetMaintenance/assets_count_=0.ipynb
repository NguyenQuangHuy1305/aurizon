```python
# Getting all possible ActivityCode from Excel file -> df1

import pandas as pd

file_path = 'MaintenanceActivityTypes.xlsx'
sheet_name = 'in'

df1 = pd.read_excel(file_path, sheet_name=sheet_name, engine='openpyxl')
df1 = df1[df1['Ignore'] != 1]

df1_split = df1.assign(TechnicalObjects_split=df1['TechnicalObjects'].str.split(', ')).explode('TechnicalObjects_split')
df1_split
```

```python
file_path = 'AssetsTypes.xlsx'
sheet_name = 'in'

asset_types_df = pd.read_excel(file_path, sheet_name=sheet_name, engine='openpyxl')

asset_types_df
```

```python
res = {col: sum(df1_split['TechnicalObjects_split'] == col) for col in asset_types_df['AssetType']}
counts = pd.DataFrame({
    'AssetType': res.keys(),
    'Count': res.values()
})
counts
# counts[counts['AssetType'] == 'Shuttle Car']
```

```python
zero_counts = counts[counts['Count'] == 0]
zero_counts
```

```python
with pd.ExcelWriter('zero_count.xlsx', mode='w', engine='openpyxl') as writer:
    # Write the DataFrame to the Excel file
    zero_counts.to_excel(writer, sheet_name='zero_count', index=False)
```
