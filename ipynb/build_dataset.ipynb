## Read in All Tonnage

-- This notebook has a few key functions:
1. Read in an existing csv called "Complete Aggregate Tonnage" which is from the TonnageFiles folder 
2.  Read in a list of entire flocs called "flocts_list.pkl" which is in the main structure
3.  Read in a csv called "Complete Conversion Linecodes" which is a csv in the main structure
    * (2) and (3) become a merged output called "flocs_joined_to_linecode"
4.  Read in "asset_renewal_all_11032024" which is the entire pkl file provided by Mike King 
    * (4) gets joined with the result for (2+3) along with (1) for tonnage 
    * result is called "nsap_renewal_assets_linked"
5.  Read in "AllFLOCMaintenanceRequirementAndAVGCost" which is the entire pkl file provided by Huy Nguyen
    * (5) gets joined with the result for (2+3) along with (1) for tonnage
    * result is called nsap_maintenance_assets_linked
    * This is the file that needs to be replace with an updated version from Huy once he can run the script that converts Jane's linkage into his maintenance format. 
    * If the format isnt clear, review the existing "AllFLOCMaintenanceRequirementAndAVGCost" file and it will become familiar as to what the format needs to be

```python
import pandas as pd
```

## 1. Read in an existing csv called "Complete Aggregate Tonnage" which is from the TonnageFiles folder 

```python
df = pd.read_csv("/workspaces/ds08-network-strategic-asset-plan-v2/TonnageFiles/Complete_Aggregate_Tonnage.csv")
df = df.rename(columns = { "DESCRIPTION":"FLOC_STORT_LocationDescr"})
df["FLOC_STORT_LocationDescr"] = df["FLOC_STORT_LocationDescr"].str.lower()
df
```

## 2.  Read in a list of entire flocs called "flocts_list.pkl" which is in the main structure

```python
df_floc = pd.read_pickle("/workspaces/ds08-network-strategic-asset-plan-v2/flocs_list.pkl")
df_floc["FLOC_STORT_LocationDescr"] = df_floc["FLOC_STORT_LocationDescr"].str.lower()
df_floc
```

```python
df_floc["FLOC_STORT_LocationDescr"] = df_floc["FLOC_STORT_LocationDescr"].str.replace(' ', '')
df_floc["FLOC_STORT_LocationDescr"] = df_floc["FLOC_STORT_LocationDescr"].str.replace('\s', '')
df_floc["FLOC_STORT_LocationDescr"] = df_floc["FLOC_STORT_LocationDescr"].str.replace('\W', '')
df_floc["FLOC_STORT_LocationDescr"] = df_floc["FLOC_STORT_LocationDescr"].str.replace('-', '')
df_floc
```

## 3.  Read in a csv called "Complete Conversion Linecodes" which is a csv in the main structure

```python
df_entire = pd.read_csv("/workspaces/ds08-network-strategic-asset-plan-v2/Conversion_Complete_LineCodes.csv", header=0)
df_entire.columns = ["System", "FLOC_STORT_LocationDescr", "LineCode"]
df_entire = df_entire.dropna()
df_entire
```

```python
df_entire["FLOC_STORT_LocationDescr"] = df_entire["FLOC_STORT_LocationDescr"].str.lower()
df_entire["FLOC_STORT_LocationDescr"] = df_entire["FLOC_STORT_LocationDescr"].str.replace(' ', '')
df_entire["FLOC_STORT_LocationDescr"] = df_entire["FLOC_STORT_LocationDescr"].str.replace('\s', '')
df_entire["FLOC_STORT_LocationDescr"] = df_entire["FLOC_STORT_LocationDescr"].str.replace('\W', '')
df_entire["FLOC_STORT_LocationDescr"] = df_entire["FLOC_STORT_LocationDescr"].str.replace('-', '')
df_entire
```

```python
complete = pd.merge(df_floc, df_entire, on = "FLOC_STORT_LocationDescr", how = 'left')
complete = complete.loc[complete["LineCode"].notnull()]
```

```python
complete.to_pickle("DataFiles/flocs_joined_to_linecode.pkl")
```

## 4.  Read in "asset_renewal_all_11032024" which is the entire pkl file provided by Mike King 
    * (4) gets joined with the result for (2+3) along with (1) for tonnage

```python
renewal = pd.read_pickle("/workspaces/ds08-network-strategic-asset-plan-v2/asset_renewal_all_11032024.pkl")
renewal = renewal.rename(columns = {"FuncLocID":"FLOC_STRNO_FunctionalLocation"})
renewal
```

```python
model_testrun_1 = pd.merge(renewal, complete, on = "FLOC_STRNO_FunctionalLocation", how = "left")
model_testrun_1.loc[model_testrun_1["LineCode"].notnull()] #this doesnt store the output so nulls still included
```

```python
model_testrun_1.to_pickle("DataFiles/nsap_renewal_assets_linked.pkl")
```
