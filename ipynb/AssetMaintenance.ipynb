```python
# Getting all FLOC from [vw_Dim_FunctionalLocation] where CompanyCode = 5000 -> df2_no_length

import pyodbc

sql_query = """
    SELECT
        FLOC_TPLNR_FunctionalLocationInternalKey,
        FLOC_STRNO_FunctionalLocation,
        FLOC_PLTXT_FunctionalLocationDescr,
        FLOC_TYPTX_FunctionalLocationCategoryDescr,
        FLOC_EQART_TechnicalObjectType,
        FLOC_EARTX_TechnicalObjectTypeDescr,
        FLOC_BUKRS_CompanyCode,
        FLOC_STORT_Location,
        FLOC_STORT_LocationDescr
    FROM [myANALYTICS_SP].[silver.dimension.enterprise.asset].[vw_Dim_FunctionalLocation]
	WHERE FLOC_BUKRS_CompanyCode = '5000'
	ORDER BY FLOC_STRNO_FunctionalLocation
"""

# Define your server name
server_name = 'myanalytics.aurizon.com.au'

# Establish a connection using Windows Authentication
conn = pyodbc.connect('DRIVER={SQL Server};SERVER=' + server_name + ';Trusted_Connection=yes;')

# Execute the SQL query and load the result into a pandas DataFrame
df2_no_length = pd.read_sql_query(sql_query, conn)

# Display the DataFrame
df2_no_length.head(10)
```

```python
# Getting all FLOC length from [vw_Dim_FunctionalLocationLinear] --> df2_length_only
import pyodbc

sql_query = """
    SELECT
        FLOCLIN_TPLNR_FunctionalLocationInternalKey,
        FLOCLIN_LINEAR_LENGTH_LinearLengthKM
    FROM [myANALYTICS_SP].[silver.dimension.enterprise.asset].[vw_Dim_FunctionalLocationLinear]
"""

# Define your server name
server_name = 'myanalytics.aurizon.com.au'

# Establish a connection using Windows Authentication
conn = pyodbc.connect('DRIVER={SQL Server};SERVER=' + server_name + ';Trusted_Connection=yes;')

# Execute the SQL query and load the result into a pandas DataFrame
df2_length_only = pd.read_sql_query(sql_query, conn)

# Display the DataFrame
df2_length_only.head(10)
```

```python
# Joining df2_no_length and df2_length_only to get df2
df2 = pd.merge(df2_no_length, df2_length_only, how='left', left_on='FLOC_TPLNR_FunctionalLocationInternalKey', right_on='FLOCLIN_TPLNR_FunctionalLocationInternalKey')

df2.head(10)  
```

```python
# Joining df1_split and df2, so that we have the all the possible Maintenance Product/Activity,... for each FLOC -> result_df
result_df = pd.merge(df2, df1_split, how='left', left_on='FLOC_EARTX_TechnicalObjectTypeDescr', right_on='TechnicalObjects_split')


## split result_df into result_df_split, join with df3 to get the average cost for each activity -> add a new AverageCost column to result_df

# Splitting result_df.ActivityCode into separate rows
result_df_split = result_df.assign(ActivityCode_split=result_df['ActivityCode'].str.split(',')).explode('ActivityCode_split')
# Joining result_df_split and df3 based on ActivityCode_split and ILART_MaintenanceActivityTypeID
merge_df2 = pd.merge(result_df_split, df3, how='left', left_on='ActivityCode_split', right_on='ILART_MaintenanceActivityTypeID')

# Calculating the average cost for each row in df1
average_cost_per_row = merge_df2.groupby(result_df_split.index)['AverageCost'].mean()

# Assigning the average cost to df1
result_df['AverageCost'] = average_cost_per_row

result_df
```
